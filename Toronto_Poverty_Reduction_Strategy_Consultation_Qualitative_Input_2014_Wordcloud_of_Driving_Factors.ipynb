```python
from wordcloud import WordCloud
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tarjin/Toronto_Open_Data/master/PRSQuestionnaireResponses%20-%20Drivers.csv')
```

```python
text = df.iloc[:, :1]
```

```python
text
```

```python
type(text)
```

```python
text = text.to_string()
```

```python
type(text)
```

```python
len(text)
```

```python
wordcloud = WordCloud(scale=2, width = 400, height=400, collocations=True, colormap='PuBu', max_words=200, background_color="black",max_font_size=40, relative_scaling=.5).generate(text)
```

```python
plt.rcParams['figure.figsize']=(20,200)
```

```python
plt.figure()
```

```python
plt.imshow(wordcloud)
```
